```swift
extends RefCounted

func create_type(state, signature):
    var type = {
        signature: {}
    }
    state["types"][signature] = type
    return ["create_type", signature]

func create_event_value(state, id, type):
    var event_value = {
        type: {}
    }
    state["events"][id] = event_value
    return ["create_event_value", id, type]

func create_event(state, id, values):
    var event = {
        values: {}
    }
    state["events"][id] = event
    return ["create_event", id, values]

func create_variable(state, type, value):
    var variable = {
        value: {}
    }
    state["variables"][type] = variable
    return ["create_variable", type, value]

func create_node_value(state, id, value, type, node=-1, socket=""):
    var node_value = {
        "value": value,
        "type": type,
        "node": node,
        "socket": socket
    }
    state["nodes"][id] = node_value
    return ["create_node_value", id, value, type, node, socket]

func create_node_configuration(state, id, value):
    var node_configuration = {
        value: {}
    }
    state["nodes"][id] = node_configuration
    return ["create_node_configuration", id, value]

func create_node_flow(state, id, node, socket):
    var node_flow = {
        "node": node,
        "socket": socket
    }
    state["nodes"][id] = node_flow
    return ["create_node_flow", id, node, socket]

func create_node(state, type, configuration=[], values=[], flows=[]):
    var node = {
        "configuration": configuration,
        "values": values,
        "flows": flows
    }
    state["nodes"][type] = node
    return ["create_node", type, configuration, values, flows]

func create_khr_interactivity(state, types=[], events=[], variables=[], nodes=[]):
    var khr_interactivity = {
        "types": types,
        "events": events,
        "variables": variables,
        "nodes": nodes
    }
    state["khr_interactivity"] = khr_interactivity
    return ["create_khr_interactivity", types, events, variables, nodes]

var state = {
    "types": {},
    "events": {},
    "variables": {},
    "nodes": {}
}
```
